### Data Exploration and Data Processing

Tahapan pertama yang kita akan lakukan adalah dengan menghitung nilai BMI dari masing-masing umur dan gender

```python
data_bmr
```

- nilai_bmr untuk laki-laki = 66 + (13.7 * Weight(kg) ) + (5 * Height(cm) ) - (6.8 * Age)
- nilai_bmr untuk perempuan = 65.5 + (9.6 * Weight(kg) ) + (1.8 * Height(cm) ) - (4.7 * Age)

```python
kolom_bmr = []

for a in range(0,len(data_bmr)):
    if data_bmr.loc[a,"Gender"] == "Male":
        nilai_bmr = 66+(13.7*data_bmr.loc[a,"Weight (kg)"])+(5*data_bmr.loc[a,"Height (cm)"])-(6.8 * data_bmr.loc[a,"Age"])
    else:
        nilai_bmr = 65.5+(9.6 * data_bmr.loc[a,"Weight (kg)"])+(1.8*data_bmr.loc[a,"Height (cm)"])-(4.7*data_bmr.loc[a,"Age"])
    kolom_bmr.append(nilai_bmr)

data_bmr["NIlai BMR"] = kolom_bmr

data_bmr
```

Dari hasil perhitungan diatas bisa dilihat nilai BMR untuk masing-masing data sudah terdapat pada kolom "Nilai BMR"
