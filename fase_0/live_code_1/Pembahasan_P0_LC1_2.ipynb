```python
#1.

def BMR_pria(BB,TB,U):
  return 66 + (13.7 * BB) + (5 * TB) - (6.8 * U)

def BMR_wanita(BB,TB,U):
  return 655 + (9.6 * BB) + (1.8 * TB) - (4.7 * U)
```

# Nomor 2

```python
#2.
import pandas as pd

data = pd.read_csv('https://raw.githubusercontent.com/rafifaditio/Dataset/main/BMR.csv')
data.head()
```

```python
bmr = []

for i in range(len(data)):
  if data['Gender'][i]=='Male':
    bmr.append(BMR_pria(data['Weight (kg)'][i],data['Height (cm)'][i],data['Age'][i]))
  elif data['Gender'][i]=='Female':
    bmr.append(BMR_wanita(data['Weight (kg)'][i],data['Height (cm)'][i],data['Age'][i]))
data['BMR'] = bmr
data.head()
```
